```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#normal distribution 
norm = np.random.normal(50, 10, 100)
norm_mean = norm.mean()
norm_std = norm.std()

# vertical lines for mean and std 
plt.axvline(norm_mean, color='r', linestyle='solid', linewidth=2)
plt.axvline(norm_mean + norm_std, color='g', linestyle='dashed', linewidth=2)
plt.axvline(norm_mean- norm_std, color='g', linestyle='dashed', linewidth=2)

#graph 
plt.hist(norm)
plt.show()


```

```python
#binomial 
binomial= np.random.binomial(100, .05, 100)


binomial_mean = binomial.mean()
binomial_std = binomial.std()

plt.axvline(binomial_mean, color='r', linestyle='solid', linewidth=2)
plt.axvline(binomial_mean + binomial_std, color='g', linestyle='dashed', linewidth=2)
plt.axvline(binomial_mean- binomial_std, color='g', linestyle='dashed', linewidth=2)

plt.hist(binomial)
plt.show()

```

```python
#poisson 
poisson= np.random.poisson(5, 100)


poisson_mean = poisson.mean()
poisson_std = poisson.std()

plt.axvline(poisson_mean, color='r', linestyle='solid', linewidth=2)
plt.axvline(poisson_mean + poisson_std, color='g', linestyle='dashed', linewidth=2)
plt.axvline(poisson_mean- poisson_std, color='g', linestyle='dashed', linewidth=2)

plt.hist(poisson, bins =10)
plt.show()
```

```python
#gamma 
gamma= np.random.gamma(5, 2,100)


gamma_mean = gamma.mean()
gamma_std = gamma.std()

plt.axvline(gamma_mean, color='r', linestyle='solid', linewidth=2)
plt.axvline(gamma_mean + gamma_std, color='g', linestyle='dashed', linewidth=2)
plt.axvline(gamma_mean- gamma_std, color='g', linestyle='dashed', linewidth=2)

plt.hist(gamma, bins =10)
plt.show()
```

```python
#geometric 
geometric= np.random.geometric(.35,100)



geometric_mean = geometric.mean()
geometric_std = geometric.std()

plt.axvline(geometric_mean, color='r', linestyle='solid', linewidth=2)
plt.axvline(geometric_mean + geometric_std, color='g', linestyle='dashed', linewidth=2)
plt.axvline(geometric_mean- geometric_std, color='g', linestyle='dashed', linewidth=2)

plt.hist(geometric ,bins = 10 )
plt.show()
```

```python
#chisquare 

chisquare =np.random.chisquare(1,100)

chisquare_mean = chisquare.mean()
chisquare_std = chisquare.std()

plt.axvline(chisquare_mean, color='r', linestyle='solid', linewidth=2)
plt.axvline(chisquare_mean + chisquare_std, color='g', linestyle='dashed', linewidth=2)
plt.axvline(chisquare_mean- chisquare_std, color='g', linestyle='dashed', linewidth=2)

plt.hist(chisquare, bins =10)
plt.show()

```

```python
pop1= np.random.normal(5,.5,100)
pop2= np.random.normal(10,1,100)
pop1_2 = pop1+pop2

plt.axvline(pop1_2.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(pop1_2.mean() + pop1_2.std(), color='g', linestyle='dashed', linewidth=2)
plt.axvline(pop1_2.mean()- pop1_2.std(), color='g', linestyle='dashed', linewidth=2)

plt.hist(pop1_2)
plt.show()
```

```python

# sorts in ascending order 
pop1_2.sort()
norm.sort()
binomial.sort()
gamma.sort()
geometric.sort()
chisquare.sort()
poisson.sort()

# graphs 
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)

plt.plot(norm,pop1_2, "o") 
plt.show() 


plt.plot(binomial,pop1_2, "o") 
plt.show() 


plt.plot(gamma,pop1_2, "o") 
plt.show() 

plt.plot(geometric,pop1_2, "o") 
plt.show() 


plt.plot(chisquare,pop1_2, "o") 
plt.show() 

plt.plot(poisson,pop1_2, "o") 
plt.show() 
```
