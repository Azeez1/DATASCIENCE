# NFL DATA 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('~/Downloads/nflstatistics/Basic_Stats.csv')
df2 = pd.read_csv('~/Downloads/nflstatistics/Career_Stats_Defensive.csv')
final = pd.DataFrame()
final['Age'] = df['Age']
final['Name'] = df['Name']
final['Weight'] = df['Weight (lbs)']
final['Experience'] = df['Experience']
final['College'] = df.College
final['Position'] = df.Position
defensive_backs_df = final[final['Position'].str.contains("CB|DB|FS|SS", na=False)] ## selects only players with the Position of CB/FS/SS/DB
defensive_lineman_df = final[final['Position'].str.contains("DT|DE",na=False)] ## selects only players with the Position of DT/DE
defensive_stats = df.merge(df2)
darrelle_revis = defensive_stats[defensive_stats['Name'].str.contains("Revis, Darrelle",na=False)] ## individual player data frame 
julius_peppers = defensive_stats[defensive_stats['Name'].str.contains("Peppers, Julius",na=False)]






```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(x = defensive_backs_df['Weight'], y = defensive_backs_df['Age'], color='green')
plt.xlabel('Weight')
plt.ylabel('Age')
plt.title('Weight  Of Defensive Backs In NFL History')


plt.subplot(1, 2, 2)
plt.scatter(x = defensive_lineman_df['Weight'], y = defensive_lineman_df['Age'], color='red')
plt.xlabel('Weight')
plt.ylabel('Age')
plt.title('Average Weight Of Defensive Lineman In NFL History')
plt.xlim(220,400)
plt.tight_layout()
plt.show()
```

```python
#PIE CHART 
labels = ['CB','FS','SS','DT','DE','LB']
freq =  [567,274,286,423,766,342]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','red','orange']

plt.pie(freq,labels = labels, colors = colors, autopct='%1.1f%%', shadow=True, startangle=140 )
plt.axis('equal')

plt.tight_layout()
plt.show()
```

```python
#BOXPLOT 
x = np.array(darrelle_revis['Solo Tackles']).astype(np.int) ## convert values to integer 
plt.boxplot(x)
plt.xlabel('Darrelle Revis')
plt.ylabel('Number Of Tackles')
plt.tight_layout()
plt.show()

```

```python
x2 = np.array(julius_peppers['Sacks']).astype(np.float) ## convert values to integer 
plt.boxplot(x2)
plt.xlabel('Julius Peppers')
plt.ylabel('Number Of sacks')
plt.tight_layout()
plt.show()
```
